Assignment 1: Dynamic Array / Array-Based List

Data Structure Implemented:
Dynamic Array (Array-Based List)

Description:
A Dynamic Array is a resizable array that automatically increases its storage capacity when full. Unlike a fixed-size array, it allows elements to be added dynamically while maintaining efficient access using indices.

Operations Implemented:
• append(value)
• get(index)
• set(index, value)
• insert(index, value)
• delete(index)
• size()

Time Complexity:

| Operation    | Complexity     |
| ------------ | -------------- |
| Access (get) | O(1)           |
| Update (set) | O(1)           |
| Append       | O(1) amortized |
| Insert       | O(n)           |
| Delete       | O(n)           |
| Size         | O(1)           |

Bioinformatics Application:
Dynamic arrays are useful for storing biological sequence data such as DNA, RNA, and protein sequences. In this assignment, the dynamic array was used to store a DNA sequence. Common genetic operations were simulated, including accessing a nucleotide, inserting a mutation, deleting a nucleotide, and replacing one nucleotide with another. Such operations are frequently performed during sequence analysis, mutation studies, and genome editing research.

Conclusion:
The DynamicArray class successfully demonstrates how a resizable array works internally and how it can be applied to bioinformatics sequence manipulation tasks.


```python
class DynamicArray:
    def __init__(self):
        self.capacity = 1
        self.count = 0
        self.data = [None] * self.capacity

    def _resize(self):
        self.capacity *= 2
        new_data = [None] * self.capacity

        for i in range(self.count):
            new_data[i] = self.data[i]

        self.data = new_data

    def append(self, value):
        if self.count == self.capacity:
            self._resize()

        self.data[self.count] = value
        self.count += 1

    def get(self, index):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of range")

        return self.data[index]

    def set(self, index, value):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of range")

        self.data[index] = value

    def insert(self, index, value):
        if index < 0 or index > self.count:
            raise IndexError("Index out of range")

        if self.count == self.capacity:
            self._resize()

        for i in range(self.count, index, -1):
            self.data[i] = self.data[i - 1]

        self.data[index] = value
        self.count += 1

    def delete(self, index):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of range")

        for i in range(index, self.count - 1):
            self.data[i] = self.data[i + 1]

        self.data[self.count - 1] = None
        self.count -= 1

    def size(self):
        return self.count

    def display(self):
        for i in range(self.count):
            print(self.data[i], end=" ")
        print()


# ==========================
# Bioinformatics Application
# ==========================

dna = DynamicArray()

sequence = ["A", "T", "G", "C", "A", "G", "T"]

for nucleotide in sequence:
    dna.append(nucleotide)

print("Original DNA Sequence:")
dna.display()

# Access nucleotide at position 3
print("\nAccess nucleotide at position 3:")
print(dna.get(2))

# Insert mutation at position 4
print("\nInsert mutation 'M' at position 4:")
dna.insert(3, "M")
dna.display()

# Delete nucleotide at position 2
print("\nDelete nucleotide at position 2:")
dna.delete(1)
dna.display()

# Replace nucleotide
print("\nReplace nucleotide at position 5 with 'T':")
dna.set(4, "T")
dna.display()

print("\nFinal Size of Dynamic Array:", dna.size())
```

Assignment 2: Singly Linked List

Data Structure Implemented:
Singly Linked List

Description:
A Singly Linked List is a linear data structure consisting of nodes. Each node stores data and a reference to the next node in the sequence. Unlike arrays, linked lists allow efficient insertion and deletion without shifting elements.

Operations Implemented:
• append(data)
• prepend(data)
• insert_after(target, data)
• delete(data)
• search(data)
• display()

Time Complexity:

| Operation    | Complexity |
| ------------ | ---------- |
| Append       | O(n)       |
| Prepend      | O(1)       |
| Insert After | O(n)       |
| Delete       | O(n)       |
| Search       | O(n)       |
| Display      | O(n)       |

Bioinformatics Application:
Singly linked lists can be used to represent contigs during genome assembly. A contig is a contiguous DNA sequence assembled from overlapping sequencing reads. In this assignment, each node represented a contig in an assembly path. Operations such as adding a new contig, removing a low-quality contig, and searching for a specific contig simulate common tasks performed during genome assembly and sequence analysis.

Conclusion:
The Singly Linked List successfully demonstrates dynamic storage and manipulation of contigs in a genome assembly pathway. Its flexibility makes it suitable for representing biological sequence relationships.


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_after(self, target, data):
        current = self.head

        while current:
            if current.data == target:
                new_node = Node(data)
                new_node.next = current.next
                current.next = new_node
                return
            current = current.next

        print("Target contig not found.")

    def delete(self, data):
        if self.head is None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return

        current = self.head

        while current.next:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next

        print("Contig not found.")

    def search(self, data):
        current = self.head

        while current:
            if current.data == data:
                return True
            current = current.next

        return False

    def display(self):
        current = self.head

        while current:
            print(current.data, end="")
            if current.next:
                print(" → ", end="")
            current = current.next

        print()


# ==========================
# Bioinformatics Application
# ==========================

assembly = SinglyLinkedList()

assembly.append("Contig1")
assembly.append("Contig2")
assembly.append("Contig3")

print("Original Genome Assembly:")
assembly.display()

# Add a new contig
print("\nAdd Contig4:")
assembly.append("Contig4")
assembly.display()

# Remove low-quality contig
print("\nRemove low-quality Contig2:")
assembly.delete("Contig2")
assembly.display()

# Search for a contig
print("\nSearch for Contig3:")
if assembly.search("Contig3"):
    print("Contig3 found.")
else:
    print("Contig3 not found.")

# Display final assembly path
print("\nFinal Assembly Path:")
assembly.display()
```

Assignment 3: Stack

Data Structure Implemented:
Stack (Last-In First-Out)

Description:
A Stack is a linear data structure that follows the Last-In First-Out (LIFO) principle. The last element inserted into the stack is the first element removed. Common stack operations include push, pop, peek, checking if the stack is empty, and determining its size.

Operations Implemented:
• push(item)
• pop()
• peek()
• is_empty()
• size()

Time Complexity:

| Operation | Complexity |
| --------- | ---------- |
| Push      | O(1)       |
| Pop       | O(1)       |
| Peek      | O(1)       |
| is_empty  | O(1)       |
| Size      | O(1)       |

Bioinformatics Application:
Stacks are commonly used in sequence alignment algorithms for traceback operations. During sequence alignment, alignment decisions such as Match, Mismatch, Insertion, and Deletion are stored while constructing the alignment matrix. During traceback, these operations are processed in reverse order to reconstruct the optimal alignment path. In this assignment, a stack was used to store alignment operations and then pop them in reverse order to simulate alignment reconstruction.

Conclusion:
The Stack implementation successfully demonstrates the LIFO principle and its application in sequence alignment traceback, an important task in bioinformatics sequence analysis.


```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items = self.items + [item]

    def pop(self):
        if self.is_empty():
            return "Stack is empty"
        
        item = self.items[-1]
        self.items = self.items[:-1]
        return item

    def peek(self):
        if self.is_empty():
            return "Stack is empty"
        
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


# ==========================
# Bioinformatics Application
# Sequence Alignment Traceback
# ==========================

alignment_stack = Stack()

print("Pushing alignment operations onto stack:")

alignment_stack.push("Match")
print("Pushed: Match")

alignment_stack.push("Match")
print("Pushed: Match")

alignment_stack.push("Insertion")
print("Pushed: Insertion")

alignment_stack.push("Match")
print("Pushed: Match")

alignment_stack.push("Deletion")
print("Pushed: Deletion")

print("\nTop Operation:")
print(alignment_stack.peek())

print("\nStack Size:")
print(alignment_stack.size())

print("\nReconstructing Alignment Path (Reverse Order):")

while not alignment_stack.is_empty():
    print(alignment_stack.pop())
```

Assignment 4: Queue

Data Structure Implemented:
Queue (First-In First-Out)

Description:
A Queue is a linear data structure that follows the First-In First-Out (FIFO) principle. The first element inserted into the queue is the first element removed. Common queue operations include enqueue, dequeue, front, checking if the queue is empty, and determining its size.

Operations Implemented:
• enqueue(item)
• dequeue()
• front()
• is_empty()
• size()

Time Complexity:

| Operation | Complexity |
| --------- | ---------- |
| Enqueue   | O(1)       |
| Dequeue   | O(n)       |
| Front     | O(1)       |
| is_empty  | O(1)       |
| Size      | O(1)       |

Bioinformatics Application:
Queues are useful for processing sequencing reads in the order they are generated by a sequencing instrument. Since reads arrive sequentially, they are naturally handled using a FIFO structure. In this assignment, sequencing reads were added to a queue and processed one by one in the same order they arrived, simulating a simple sequencing analysis workflow.

Conclusion:
The Queue implementation successfully demonstrates FIFO processing and its application in handling sequencing reads, which is a common task in bioinformatics pipelines.


```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items = self.items + [item]

    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"

        item = self.items[0]
        self.items = self.items[1:]
        return item

    def front(self):
        if self.is_empty():
            return "Queue is empty"

        return self.items[0]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


# ==========================
# Bioinformatics Application
# Sequencing Read Processing
# ==========================

read_queue = Queue()

reads = ["Read1", "Read2", "Read3", "Read4"]

print("Enqueuing Sequencing Reads:")

for read in reads:
    read_queue.enqueue(read)
    print(f"Enqueued: {read}")

print("\nFirst Read in Queue:")
print(read_queue.front())

print("\nQueue Size:")
print(read_queue.size())

print("\nProcessing Reads in Arrival Order:")

while not read_queue.is_empty():
    print(read_queue.dequeue())
```
